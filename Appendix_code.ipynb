### 1.0.3. Functions


```python
def collect_tweets (list_of_usernames, list_of_tweets, max_number = 3200):
    """
    Function for collecting the last max_number tweets for each politician in a list
    Input
    max_number: number of tweets for each account (max 3200)
    list_of_usernames: list of twitter handles without @
    list_of_tweets: list to save the information


    """
    # create, then update a .json file
    for i in tqdm(list_of_usernames):
        # define user
        user_metrics = client.get_user(username = i, user_fields = ['public_metrics'])
        user_id = user_metrics['data']['id']
        user_number_of_tweets = int(user_metrics['data']['public_metrics']['tweet_count'])

        # get last max_number tweets
        all_user_tweets = []
        while (len(all_user_tweets) < max_number) & (len(all_user_tweets) < user_number_of_tweets):
            if len(all_user_tweets) == 0:

                # getting the first 100 tweets
                tweets = client.get_users_tweets(user_id,
                                                 tweet_fields = ['created_at', 'public_metrics', 'referenced_tweets', 'text'],
                                                 max_results = 100)
                all_user_tweets += tweets['data']

            else:
                # getting subsequent groups of 100 tweets
                last_tweet = all_user_tweets[-10]['id']
                tweets = client.get_users_tweets(user_id,
                                                 tweet_fields = ['created_at', 'public_metrics', 'referenced_tweets', 'text'],
                                                 max_results = 100, until_id = last_tweet)
                all_user_tweets += tweets['data']

        # create dict
        user_ids = []
        username = []
        text = []
        tweet_id = []
        created_at = []
        retweeted = []
        original_tweet_id = []
        retweets = []
        replies = []
        likes = []
        impression = []

        for tweet in all_user_tweets:

            user_ids.append(user_id)
            username.append(i)
            text.append(tweet['text'])
            tweet_id.append(tweet['id'])
            created_at.append(tweet['created_at'])
            retweets.append(tweet['public_metrics']['retweet_count'])
            replies.append(tweet['public_metrics']['reply_count'])
            likes.append(tweet['public_metrics']['like_count'])
            impression.append(tweet['public_metrics']['impression_count'])

            if ('referenced_tweets' in tweet):
                retweeted.append(1)
                original_tweet_id.append(tweet['referenced_tweets'][0]['id'])
            else:
                retweeted.append(0)
                original_tweet_id.append(0)

        list_of_tweets.append({'id': user_id,
                            'username': username,
                            'text': text,
                            'tweet_id': tweet_id,
                            'created_at': created_at,
                            'retweeted': retweeted,
                            'original_tweet_id': original_tweet_id,
                            'retweets': retweets,
                            'replies': replies,
                            'likes': likes,
                            'impression': impression})
```

```python
#Reusing code from class
def preprocess_no_stopwords(text):

    # remove bold formatting, not working
    text = re.sub(r'\*{2}(.*?)\*{2}', r'\1', text)
    # remove italic formatting, not working
    text = re.sub(r'_{1}(.*?)_{1}', r'\1', text)

    #Lowercasing words
    text = text.lower()

    #Removing '&amp' which was found to be common
    text = re.sub(r'&amp','', text)

    #Replace other instances of "&" with "and"
    text = re.sub(r'&','and', text)

    #Removing mentions
    text = re.sub(r'@\w+ ', '', text)

    #Removing 'RT' and 'via'
    text = re.sub(r'(^rt|^via)((?:\b\W*@\w+)+): ', '', text)

    #Removing punctuation
    my_punctuation = string.punctuation.replace('#','') # We don't want "#" or "-" removed
    my_punctuation = my_punctuation.replace('-','')
    my_punctuation = my_punctuation+'’“”—'  # Adding additional punctuation that is not captured by string.punctuation

    text = text.translate(str.maketrans('', '', my_punctuation)) # Text is "translated" to not include the punctuation we defined

    #Removing odd special characters
    text = re.sub(r' - ','', text) #removing dash lines bounded by whitespace (and therefore not part of a word)
    text = re.sub(r"[┻┃━┳┓┏┛┗]","", text)
    text = re.sub(r"\u202F|\u2069|\u200d|\u2066","", text)

    #Removing URLs
    text = re.sub(r'http\S+', '', text)

    #Removing numbers
    text = re.sub(r'\d+','', text)

    patterns = set(stopwords.words('english')) # Removing stopwords

    for pattern in patterns:
        if re.search('\\b'+pattern+'\\b', text):           #Searching for exact match of stopwords in each tweet
            text = re.sub('\\b'+pattern+'\\b', '', text)   #Substituting stopwords with empty string
    text = re.sub(r'\s+',' ',text)                           #Removing stopwords creates whitespace which we remove here
    text = text.strip()


    #Tokenizing (separators and superfluous whitespace are also removed during tokenization)
    tokenizer = TweetTokenizer()
    tokens = tokenizer.tokenize(text)
    return tokens
```

```python
def preprocess(text):

    # remove bold formatting, not working
    text = re.sub(r'\*{2}(.*?)\*{2}', '', text)
    # remove italic formatting, not working
    text = re.sub(r'_{1}(.*?)_{1}', '', text)

    #Lowercasing words
    text = text.lower()

    #Removing '&amp' which was found to be common
    text = re.sub(r'&amp','', text)

    #Replace other instances of "&" with "and"
    text = re.sub(r'&','and', text)

    #Removing mentions
    text = re.sub(r'@\w+ ', '', text)

    #Removing 'RT' and 'via'
    text = re.sub(r'(^rt|^via)((?:\b\W*@\w+)+): ', '', text)

    #Removing punctuation
    #Using the module string to define what punctuation we want removed
    my_punctuation = string.punctuation.replace('#','') # We don't want "#" or "-" removed
    my_punctuation = my_punctuation.replace('-','')
    my_punctuation = my_punctuation+'’“”—'  # Adding additional punctuation that is not captured by string.punctuation

    text = text.translate(str.maketrans('', '', my_punctuation)) # Text is "translated" to not include the punctuation we defined

    #Removing odd special characters
    text = re.sub(r' - ','', text) #removing dash lines bounded by whitespace (and therefore not part of a word)
    text = re.sub(r"[┻┃━┳┓┏┛┗]","", text)
    text = re.sub(r"\u202F|\u2069|\u200d|\u2066","", text)

    #Removing URLs
    text = re.sub(r'http\S+', '', text)

    #Removing numbers
    text = re.sub(r'\d+','', text)

    #Tokenizing (separators and superfluous whitespace are also removed during tokenization)
    tokenizer = TweetTokenizer()
    tokens = tokenizer.tokenize(text)

    return tokens
```

```python
def co_occurance(size_embedding = 300, window_context = 10, min_count_ = 1, party = 'Conservative', top_words = 20,
                 word_ = 'inflation',
                 df = pd.DataFrame({})):
  """
  Word2Vec condition on political party
  """
  text_everything = df[df['Party']==party]['tokens_no_stopwords'].to_list()
  model_word2vec = Word2Vec(text_everything, vector_size=size_embedding,
                    window=window_context, min_count=min_count_, workers=4)
  list_cooccurance_words = [x[0] for x in model_word2vec.wv.most_similar(word_, topn=top_words)]
  return list_cooccurance_words
```

```python
# merge to have the same key as nodes from the network
appeal_use_name = pd.merge(user_data_df[['real_name','username']],
         appeal_use[['username', 'pathos_negative_mean', 'pathos_sympathy_mean', 'logos_mean','ethos_mean']])

# assign the mean of each appeal to attributes of the network
# appeal 1
username_pathos_negative_dict = {}
for index, row in appeal_use_name.iterrows():
    username = row['real_name']
    mean = row['pathos_negative_mean']
    username_pathos_negative_dict[username] = mean
nx.set_node_attributes(G, username_pathos_negative_dict, "pathos_negative_mean")
# appeal 2
username_pathos_sympathy_dict = {}
for index, row in appeal_use_name.iterrows():
    username = row['real_name']
    mean = row['pathos_sympathy_mean']
    username_pathos_sympathy_dict[username] = mean
nx.set_node_attributes(G, username_pathos_sympathy_dict, "pathos_sympathy_mean")
# appeal 3
username_logos_dict = {}
for index, row in appeal_use_name.iterrows():
    username = row['real_name']
    mean = row['logos_mean']
    username_logos_dict[username] = mean
nx.set_node_attributes(G, username_logos_dict, "logos_mean")
# appeal 4
username_ethos_dict = {}
for index, row in appeal_use_name.iterrows():
    username = row['real_name']
    mean = row['ethos_mean']
    username_ethos_dict[username] = mean
nx.set_node_attributes(G, username_logos_dict, "ethos_mean")

# exporing graph for visualisation in Gephi
name_network = '20230608_rts_with_node_attrs.gexf'
nx.write_gexf(G, os.path.join(output_path, name_network))
```
